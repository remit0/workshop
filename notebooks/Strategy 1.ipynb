# what we need
- assign large families first -> family cost function is dependent on family size
- identify families that are "tolerant" and those are not. (-> tolerant: diverse day choices)
- assign large and intolerant families first
- identify days that should be kept to a minimum (right before the weekend / holidays for acc penalty) + unwanted ones
- secure min constraint first vs redistributing

```python
data = get_data()
maker = init(data)
```

```python
for day in range(1, 101):
    groups_scored = maker.evaluate_groups(day)
    for group, score in groups_scored:
        maker.assign(group, day)
        if maker.bookings.loc[day] >= maker.MIN:
            break
```

```python
def get_best_wish(maker, group):
    valid_days = maker.bookings[maker.bookings <= maker.MAX - group.size].index.tolist()
    days = [day for day in valid_days if day in group.days]
    if not days:
        days = [day for day in valid_days if day not in group.days]
    chosen_day, min_cost = -1, float("inf") # pas super élégant mais permet de ne pas appeler sorted sinon ça prend 3 heures
    for day in days:
        day_score = maker.evaluate_one_group(group, day)
        if day_score < min_cost:
            min_cost = day_score
            chosen_day = day
    return chosen_day
```

```python
for group in maker.groups:
    day = get_best_wish(maker, group)
    maker.assign(group, day)

maker.make_submission("strat1")
```
